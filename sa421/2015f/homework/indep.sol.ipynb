Use `rand()` from `numpy.random` to generate 100 random numbers ([documentation](http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.rand.html)).
Plot the lag-$k$ autocorrelation for $k = 0,1,\dots,50$.
Based on these autocorrelation values, do these random numbers appear to be independent?

```python
##### Setup #####
# Import random from numpy.random
from numpy.random import rand

# Import pearsonr from scipy.stats
from scipy.stats import pearsonr

# Import plot from Matplotlib
from matplotlib.pyplot import plot

# Run Matplotlib magic to show plots directly in the notebook
%matplotlib inline

# Make Matplotlib plots display as SVG files, which are cleaner
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
# Create list of 100 random numbers
R = [rand() for i in range(100)]

# Get number of observations/random numbers
n = len(R)

# Compute lag-k autocorrelation for k = 0,1,...,50
lagAC = []
for k in range(51):
    [corr, pvalue] = pearsonr(R[0:n-k], R[k:n])
    lagAC.append(corr)

# Plot the lag-k autocorelation for k = 0,1,...,50
plot(lagAC, 'o')
```

The autocorrelation values are generally between -0.3 and 0.3, indicating relatively small amounts of autocorrelation. Based on our rule of thumb, we do not reject the hypothesis that these random numbers are independent.

# Do the tweets in my Twitter feed follow a Poisson process?

Let's use R to perform goodness-of-fit tests on the interarrival times of tweets in my Twitter feed.

We can call R in IPython Notebook by using the R magic extension.
Let's start by loading it.

```python
%load_ext rmagic
```

We'll need the `fitdistrplus` library for R. 

Note: there are two ways to use R commands. We can start a block with `%%R` on the first line - then the entire block will be executed as R code. We can also execute a single line as R code by starting the line with `%R`.

```python
%R library(fitdistrplus)
```

Now, let's read in the data from `twitter.txt`.

```python
%%R
twitter <- read.table('twitter.txt')
arrival <- twitter$V1
```

The tweet arrival times are formatted as YYYY-MM-DDTHH:MM:SSZ, so we need to tell R this. This will convert the dates into the number of seconds since Thursday 1 January 1970, 0:00 UTC.

```python
%%R 
arrival <- as.POSIXct(arrival, "%Y-%m-%dT%H:%M:%SZ", tz="UTC")
```

Next, let's compute the interrival times by

1. sorting the arrival times, and 
2. computing the difference in consecutive arrival times.

We divide the interarrival times by 60 to obtain times in minutes.

```python
%%R
# Sort by arrival time
sort(arrival)

# Compute interarrival times in minutes
interarrival <- as.numeric(diff(arrival)) / 60
```

The MLE estimator for the exponential distribution is the average arrival rate. Let's compute this next.

```python
%%R
rate <- 1 / mean(interarrival)
message("Average interarrival rate")
print(rate)
```

Next, let's use `fitdist` from the `fitdistrplus` library to determine how well the interarrival times fit with the exponential distribution..

```python
%%R
# Goodness of fit 
fit <- fitdist(interarrival, distr="exp")

# Get goodness-of-fit statistics
gof <- gofstat(fit)

# Display fit and gof 
print(fit)
print(gof)
message("K-S test reject H0 at significance level 0.05?")
print(gof$kstest)
message("CvM test reject H0 at significance level 0.05?")
print(gof$cvmtest)
message("A-D test reject H0 at significance level 0.05?")
print(gof$adtest)
```

Let's check visually what the fit looks like. Here are comparisons of the empircal cdf and pdf with their theoretical counterparts.

```python
%%R
# Get cdf comparison
cdfcomp(fit, legendtext=c("exponential"), xlab="time (min)", ylab="cdf")

# Get pdf comparison
denscomp(fit, legendtext=c("exponential"), xlab="time (min)", ylab="density")
```

We also need to check independence of the interarrival times. One easy visual test is to plot the interarrival times as a time series.

```python
%%R 
plot(interarrival, type="o", col="blue", xlab="arrival (tweet)", ylab="interarrival time (min)", main="Interarrival times")
```

We can also look at autocorrelations:

```python
%%R 
acf(interarrival, xlab="lag between arrivals", ylab="autocorrelation", main="Autocorrelation")
```
